```python
import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
cards = pd.read_json (r'data/cards.json')
cards = json_normalize(cards['data'].values)
cards.to_pickle('cards.pkl')
pd.read_pickle('cards.pkl')
```

```python
df =  pd.read_json(r'data/decks.json')
deck_list = pd.concat([json_normalize(x)['data'] for x in df.values])
mask = deck_list.map(lambda d: len(d)) > 0
deck_list = deck_list[mask]
decks = pd.concat([json_normalize(x) for x in deck_list.values],sort=True,ignore_index=True)
decks
```

```python
decks.to_pickle('decks.pkl')
```
